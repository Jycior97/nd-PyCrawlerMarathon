```python
#1. 取出 台北市士林區 2018/01 – 2018/08 的 SO2 資料
```

```python
from selenium import webdriver
from selenium.webdriver.support.ui import Select
from bs4 import BeautifulSoup
```

```python
browser = webdriver.Chrome(executable_path='chromedriver')

browser.get("http://taqm.epa.gov.tw/taqm/tw/MonthlyAverage.aspx")
```

```python
selectSite = Select(browser.find_element_by_id("ctl05_ddlSite"))
selectSite.select_by_value('11') #透過開發者工具(F12) 查找
selectYear = Select(browser.find_element_by_id("ctl05_ddlYear"))
selectYear.select_by_value('2018')

browser.find_element_by_id('ctl05_btnQuery').click()
```

```python
html_source = browser.page_source

soup = BeautifulSoup(html_source, 'html.parser')
table = soup.find('table', class_='TABLE_G')
print(table)
```

```python
value = []
i = 0
while True:
    value.append({
        '月份:': table.find_all('td')[2+i*3].text,
        'SO2(ppb):': table.find_all('td')[3+i*3].text})
    i+=1
    if i == 7 :
        break

import pandas as pd
pd.DataFrame(value)
```

```python
# 2. 取出 台北市士林區 2018/01 – 2018/08 的 SO2、CO 資料
```

```python
for i in range(0,len(table.find_all('td'))):
    if table.find_all('td')[i].text == 'CO' :
        print('CO is at :',i)
        position_CO = i
    if table.find_all('td')[i].text == 'SO2' :
        print('SO2 is at :',i)
        position_SO2 = i
    else :
        pass
```

```python
value = []
i = 0
SO2_ms = table.find_all('td')[position_SO2+1].text
CO_ms = table.find_all('td')[position_CO+1].text
while True:
    value.append({
        '月份:': table.find_all('td')[2+i*3].text,
        'SO2('+SO2_ms+'):': table.find_all('td')[3+i*3].text,
        'CO('+CO_ms+'):':table.find_all('td')[position_CO+2+i*3].text})
    i+=1
    if i == 7 :
        break

import pandas as pd
pd.DataFrame(value)
```
