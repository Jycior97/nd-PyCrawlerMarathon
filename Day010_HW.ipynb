```python
#requests + BeautifulSoup

import requests
from bs4 import BeautifulSoup
url = 'https://www.dcard.tw/f'
r = requests.get(url)
r.encoding = 'utf-8'
soup = BeautifulSoup(r.text, "html5lib")

print(soup)
```

```python
#Grab + Pyquery
```

```python
from grab import Grab
g = Grab()
resp = g.go(url) 
resp.body 
```

```python
from pyquery import PyQuery as pq
doc = pq(resp.body)
title = doc('title')
title.text
```

```python
print(type(title), title.text())
```
